```python
import pygame
import sys
import time
import random

# Configurações de Dificuldade do jogo
difficulty = 25

# Dimensões da tela do game
frame_size_x = 720
frame_size_y = 480

# Inicializando o Pygame
check_errors = pygame.init()
if check_errors[1] > 0:
    print(f'[!] Houve {check_errors[1]} erros ao inicializar o jogo, saindo...')
    sys.exit(-1)
else:
    print('[+] Jogo inicializado com sucesso')

# Inicializando a tela do game
pygame.display.set_caption('Snakegame')
game_window = pygame.display.set_mode((frame_size_x, frame_size_y))

# Cores (R, G, B)
black = pygame.Color(0, 0, 0)
white = pygame.Color(255, 255, 255)
red = pygame.Color(255, 0, 0)
green = pygame.Color(0, 255, 0)
blue = pygame.Color(0, 0, 255)

# Controle de FPS (frames por segundo)
fps_controller = pygame.time.Clock()

# VariÃ¡veis do jogo
snake_pos = [100, 50]
snake_body = [[100, 50], [100-10, 50], [100-(2*10), 50]]
snake_direction = 'RIGHT'  # Direção inicial da cobra

food_pos = [random.randrange(1, (frame_size_x//10)) * 10, random.randrange(1, (frame_size_y//10)) * 10]
food_spawn = True

# Pontuação do jogo
score = 0

# Função para imprimir "Game Over" e score
def game_over():
    print(f"Game Over! score: {score}")
    pygame.quit()
    sys.exit()

# Main logic
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_UP and not snake_direction == 'DOWN':
                snake_direction = 'UP'
            elif event.key == pygame.K_DOWN and not snake_direction == 'UP':
                snake_direction = 'DOWN'
            elif event.key == pygame.K_LEFT and not snake_direction == 'RIGHT':
                snake_direction = 'LEFT'
            elif event.key == pygame.K_RIGHT and not snake_direction == 'LEFT':
                snake_direction = 'RIGHT'

    # Logica do jogo Snake
    if snake_direction == 'UP':
        snake_pos[1] -= 10
    if snake_direction == 'DOWN':
        snake_pos[1] += 10
    if snake_direction == 'LEFT':
        snake_pos[0] -= 10
    if snake_direction == 'RIGHT':
        snake_pos[0] += 10

    # Verificando as bordas e a cobra tocando nela mesma
    if (
        snake_pos[0] < 0
        or snake_pos[0] > frame_size_x - 10
        or snake_pos[1] < 0
        or snake_pos[1] > frame_size_y - 10
    ):
        game_over()

    # Verificando se a cobra toca na maçã
    if snake_pos[0] == food_pos[0] and snake_pos[1] == food_pos[1]:
        food_spawn = True
        score += 1
    else:
        snake_body.pop()

    # Movendo a cobra
    snake_body.insert(0, list(snake_pos))

    # Spawn 
    if food_spawn:
        food_pos = [random.randrange(1, (frame_size_x//10)) * 10, random.randrange(1, (frame_size_y//10)) * 10]
        food_spawn = False

    # Desenhando a tela do jogo
    game_window.fill(black)

    # Desenhando a maçã vermelha
    pygame.draw.rect(game_window, red, pygame.Rect(food_pos[0], food_pos[1], 10, 10))

    # Desenhando a cobra verde
    for pos in snake_body:
        pygame.draw.rect(game_window, green, pygame.Rect(pos[0], pos[1], 10, 10))

    # Atualizando a tela do jogo
    pygame.display.update()
    # Taxa de atualização
    fps_controller.tick(difficulty)
```
